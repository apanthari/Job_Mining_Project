```python
#convert all values to yearly
def salary_conversion_lower(row):
    type = str(row['type']) 
    lower_wage = row['lower_wage']
    #upper_wage

    if type == 'hourly':
        return lower_wage*2080
    elif type == 'weekly':
        return lower_wage*52
    elif type == 'monthly':
        return lower_wage*12
    else: 
        return lower_wage

# Apply the labeling function
manual_labeling_anuja['converted_yearly_lower'] = manual_labeling_anuja.apply(salary_conversion_lower, axis=1)
```

```python
manual_labeling_anuja
```

```python
#convert all values to yearly
def salary_conversion_upper(row):
    type = str(row['type']) 
    upper_wage = row['upper_wage']
    #upper_wage

    if type == 'hourly':
        return upper_wage*2080
    elif type == 'weekly':
        return upper_wage*52
    elif type == 'monthly':
        return upper_wage*12
    else: 
        return upper_wage

# Apply the labeling function
manual_labeling_anuja['converted_yearly_upper'] = manual_labeling_anuja.apply(salary_conversion_upper, axis=1)
```

##### Manually label jobs

```python
labeled = manual_labeling_anuja.copy()
```

```python
# Define keywords for job levels
entry_keywords = ['integrity', 'programmer', 'statistical', 'statistician', 'collection', 'admin', 'agent', 'early', 'analytics', ' 2', 'designer', 'administrator', ' i ','staff', 'dba', 'junior', 'assistant', 'entry', 'graduate', 'intern', 'trainee', 'aide','collector','technician' ]
mid_keywords = ['warehousing','migration', 'migrations', 'big data', 'machine learning', 'ai', 'data scientist', 'steward', ' ii','mid', 'associate', 'specialist', 'coordinator', 'analyst', 'engineer', 'developer']
senior_keywords = ['advisor', 'founding', 'product', 'vp,', 'vp ', 'architect', 'mentor', 'supervisor', 'sr. ','head ','sr ', 'president', ' iii','senior', 'lead', 'manager', 'director', 'principal', 'executive', 'chief']

# Function to label job levels based on the title and blurb
def label_job_level(row):
    title = str(row['title']).lower()
    blurb = str(row['blurb']).lower()
    
    # if job_level == 'Entry-Level' or job_level == 'Mid-Level' or job_level == 'Senior-Level':
    #     return str(row('job_level'))
    if any(keyword in title or keyword in blurb for keyword in senior_keywords):
        return 'Senior-Level'
    elif any(keyword in title or keyword in blurb for keyword in mid_keywords):
        return 'Mid-Level'
    elif any(keyword in title or keyword in blurb for keyword in entry_keywords):
        return 'Entry-Level'
    else:
        return 'Uncategorized'

# Apply the labeling function
labeled['job_level'] = labeled.apply(label_job_level, axis=1)

# Count job levels for an overview
job_level_counts = labeled['job_level'].value_counts()
job_level_counts
```
