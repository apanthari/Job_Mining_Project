```python
import pandas as pd
```

```python
jobs = pd.read_csv('all_jobs.csv')
jobs.head()
```

```python
jobs['state_zip'] = jobs['location'].str.split(',').str[1]
```

```python
jobs.head()
```

```python
#Bar chart of count of jobs in each state
jobs['state'] = jobs['state_zip'].str.split(' ').str[1]
```

```python
jobs.head()
```

```python
# Count each value in the 'category' column
counts = jobs['state'].value_counts()

# Create the bar chart
counts.plot.bar(title='Value Counts')
```

```python
#count of jobs in each city by state
```

```python
jobs['city'] = jobs['location'].str.split(',').str[0]
```

```python
jobs
```

```python
ca = jobs.loc[jobs['state'] == 'CA']
ca
```

```python
# Count each value in the 'category' column
counts = ca['city'].value_counts()

# Create the bar chart
counts.plot.bar(title='Value Counts')
```

```python
#query every row that has san francisco
sf = ca[ca['city'].str.contains('San Fran')]
sf['city'].value_counts()
```
