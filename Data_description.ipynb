```python
import pandas as pd
```

```python
jobs = pd.read_csv('all_jobs.csv')
```

```python
#create state column
jobs['state_zip'] = jobs['location'].str.split(',').str[1]
jobs['state'] = jobs['state_zip'].str.split(' ').str[1]
#create city column
jobs['city'] = jobs['location'].str.split(',').str[0]
```

```python
# Count each value in the 'category' column
counts_state = jobs['state'].value_counts()

# Create the bar chart
counts_state.plot.bar(title='Value Counts')
```

```python
#count of jobs in each city by state
```

```python
ca = jobs.loc[jobs['state'] == 'CA']
counts_ca = ca['city'].value_counts()
```

```python
top_10_ca = counts_ca.to_frame().head(10)
top_10_ca.plot.bar(title='Cities in CA')
```
