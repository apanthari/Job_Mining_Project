```python
import pandas as pd
manual_labeling_anuja = pd.read_csv('crowdsource_anuja.csv')
manual_labeling_anuja['title'] = manual_labeling_anuja['title'].str.lower()
```

```python
labeled = manual_labeling_anuja.copy()
```

```python
# Define keywords for job levels
entry_keywords = [' 2', 'designer', 'administrator', ' i ','staff', 'dba', 'junior', 'assistant', 'entry', 'graduate', 'intern', 'trainee', 'aide','collector','technician' ]
mid_keywords = ['data scientist', 'steward', ' ii','mid', 'associate', 'specialist', 'coordinator', 'analyst', 'engineer', 'developer']
senior_keywords = ['vp ', 'architect', 'mentor', 'supervisor', 'sr. ','head ','sr ', 'president', ' iii','senior', 'lead', 'manager', 'director', 'principal', 'executive', 'chief']

# Function to label job levels based on the title and blurb
def label_job_level(row):
    title = str(row['title']).lower()
    blurb = str(row['blurb']).lower()
    #print(str(row['job_level']))
    
    # if job_level == 'Entry-Level' or job_level == 'Mid-Level' or job_level == 'Senior-Level':
    #     return str(row['job_level'])
    if any(keyword in title or keyword in blurb for keyword in senior_keywords):
        return 'Senior-Level'
    elif any(keyword in title or keyword in blurb for keyword in mid_keywords):
        return 'Mid-Level'
    elif any(keyword in title or keyword in blurb for keyword in entry_keywords):
        return 'Entry-Level'
    else:
        return 'Uncategorized'

# Apply the labeling function
labeled['job_level'] = labeled.apply(label_job_level, axis=1)

# Count job levels for an overview
job_level_counts = labeled['job_level'].value_counts()
job_level_counts
```
